```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
def get_loggs(name):
    return pd.read_csv(Path('../exp/') / name / 'history.csv', index_col=0)['mae']
    
    
def plot_logs(experiment_names, xlabel='Step', ylabel='Mean Absolute Error', title=''):
    """
    Plots training curves for different logs based on MAE with a specific aesthetic.

    Args:
    experiment_names : list of str
        List containing the names of the experiments.
    xlabel : str
        Label for the x-axis.
    ylabel : str
        Label for the y-axis.
    title : str
        Title of the plot.

    Returns:
    matplotlib.figure.Figure
        The figure object with the plot.
    """
    
    # Colors for the plot lines. You can adjust them as per your requirement.
    colors = ['#2c3e50', '#e74c3c', '#3498db']
    
    # Create a new figure and set the size
    fig, ax = plt.subplots(figsize=(12, 6))

    for i, exp_name in enumerate(experiment_names):
        log = get_loggs(exp_name)
        
        epochs = log.index.values
        mae = log.values  # MAE values from logs

        ax.plot(epochs, mae, color=colors[i], label=exp_name, lw=2.5, alpha=0.5)
    
    # Set the title and labels
    ax.set_title(title, fontsize=16, fontweight='bold', y=1.05)
    ax.set_xlabel(xlabel, fontsize=14, fontweight='bold')
    ax.set_ylabel(ylabel, fontsize=14, fontweight='bold')

    # Legend, grid and layout settings
    ax.legend(loc='upper right', frameon=True, framealpha=1, fontsize=12, edgecolor='black')
    ax.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()

    # Adjusting spines (the borders) to make it look cleaner
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_color('#333333')
    ax.spines['left'].set_color('#333333')
    
    ax.tick_params(colors='#333333')

    # Show the plot
    plt.show()



```

```python
experiment_names = ['exp_40', 'exp_32']
plot_logs(experiment_names)
```
